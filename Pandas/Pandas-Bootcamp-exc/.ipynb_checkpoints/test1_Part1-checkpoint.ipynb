### Best and worst practices and limitations

```python
titanic.loc[1,'age'] =40 #change the row label in age column to be 40, even in original df.
```

```python
titanic.iloc[1, 3] =41 #iloc works just as fine.
```

```python
titanic.loc[1:3, 'age'] =42
```

```python
titanic.iloc[1:4,3] = [32,43,47] #multiple assignment
```

```python
index_babies = titanic.loc[titanic.age<1,'age'].index
```

```python
index_babies
```

```python
titanic.loc[titanic.age<1,'age'] = 1
```

```python
titanic.loc[index_babies]
```

```python
titanic.head(4)
```

```python
titanic.loc[0,'survived':'sex'] = [1,3,'female'] #overwrite values indivisually.
```

```python
titanic.replace(0,'zero')#replaces values.
```

```python
age=titanic.age #1st indexing
```

```python
age[1]= 40 #2nd indexing a.k.a chain-indexing. Hence, the error.
```

```python
titanic.age[1] =41 #again chain indexing.
```

Use loc operator as norm. Other techniques are not reliable.

```python
titanic.loc[1, 'age'] = 42 # See!! No error with loc.
```

```python
slice1 = titanic[['age', 'sex']]#chain 1
```

```python
slice1.iloc[1,0] = 47 # chain 2. still chain indexing. original df remains unchained.
```

```python
slice2=titanic.loc[:,['sex','age']]#chain1
```

```python
slice2.iloc[1,1] = 41#chain2. Original df remains unchanged.
```
