#### Limitations of pivot

```python
table2 = pd.read_csv('table2.csv')
```

```python
table2
```

```python
# duplicates in table2.
#pivot() method is not designed for multi-index.
table2.pivot(index = 'Country', columns= 'Medal', values = 'Count')
#confused with duplicate locations.
```

```python
table2.groupby(['Country', 'Year', 'Medal']).Count.sum().unstack(fill_value=0)
```

#### pivot table

```python
table2.pivot_table(index= 'Country', columns='Medal', values= 'Count',#default aggfunc is 'mean'
                   aggfunc= 'sum',fill_value=0,margins=True,margins_name='Total')
```

```python
table2.groupby(['Country','Medal']).Count.sum().unstack().fillna(0)#same as pivot_table
#no built in func margins for groupby
```

```python
agg =table2.groupby(['Country','Medal']).Count.sum().unstack().fillna(0)
```

```python
agg['Total']= agg.sum(axis = 1)#for column. 
```

```python
agg.loc['Total']= agg.sum(axis=0)#for row.
```

```python
agg
```

### pd.crosstab()

```python
#cross tabulations for frequencies, that is count of one respct to other.
pd.crosstab(titanic.sex, titanic.pclass, margins=True,normalize=True)
```

```python
titanic.groupby(['sex','pclass']).pclass.count().unstack()
```

```python
#pd.crosstab() additional functionalities.
#titanic.sex and titanic.pclass can be from different df.
pd.crosstab(titanic.sex, titanic.pclass, values = titanic.age, aggfunc='mean')
```

```python
#above can be achieved with pivot_table with 'age' values.
#one titanic df is applied to pivot_table. Limited to one df.
titanic.pivot_table(index='sex',columns='pclass', values= 'age',aggfunc='mean')
```

```python
titanic.groupby(['sex','pclass']).age.mean().unstack()
```

```python
#another example
pd.crosstab(titanic.sex, titanic.pclass, values = titanic.fare, 
            aggfunc='sum',margins=True,normalize=True)
```
